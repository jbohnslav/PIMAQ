```python
import os
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
file = r'D:\DATA\JB\realsense\experiment08\front_right_JB043.h5'
assert(os.path.isfile(file))
```

```python
with h5py.File(file, 'r') as f:
    times = f['time'][:]
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(times)
ax[1].plot(np.diff(times)*1000)
# plt.axis('square')
plt.show()
```
